```python
# create function to combine 2 file and sorting the lyric in file 
import os
def combine_sort_file(file1,file2,newfile='imagine.txt'):
  """
  This function combine 2 txt file and sorting
  file1, file2 = files to be merged
  newfile = the new file name that has been sorted (default:imagine.txt)
  """
  with open(file1) as f:
    lines1 = [line.rstrip() for line in f]

  with open(file2) as f:
    lines2 = [line.rstrip() for line in f]

  lyric = lines1 + lines2

  dict_lyric ={}
  for line in lyric:
      lines = line.split()
      num = lines[0].replace('.','')
      value_lyric = ' '.join(lines[1:])
      (key, val) = num,value_lyric
      dict_lyric[int(key)] = val

  with open(newfile, 'w') as f:
    for key in sorted(dict_lyric):
      f.write(str(key) + '. ' + dict_lyric[key]+'\n')

  file = open(newfile,'r')
  readfile = file.read()
  print(readfile)
  print('File created successfully')
  print('File path:\n  ',os.path.abspath(newfile))
```

```python
im1 = "Imagine/imagine-1.txt"
im2 = "Imagine/imagine-2.txt"
combine_sort_file(im1, im2, 'bayangkan_sorted.txt')
```

```python
# creat function that used for reversing the lines and words in each line from the file
def reverse_lyric(file1, filename = 'Imagine-double-reversed.txt'):
  """
  This function is used for reversing the lines and words in each line from the file
  file1 = the file to be reversed
  filename = filename that has been reversed(default : Imagine-double-reversed.txt)
  """
  with open(file1) as f:
    lines_imag = [line.rstrip() for line in f]
  lines_imag.reverse()

  with open(filename, 'w') as f:
     for line in lines_imag:
         lines = line.split()
         lines.reverse()
         reverseline = ' '.join(lines)
         f.write(reverseline + '\n')

  file = open(filename,'r')
  readfile = file.read()
  print(readfile)
  print('File created successfully')
  print('File path:\n  ',os.path.abspath(filename))
```
